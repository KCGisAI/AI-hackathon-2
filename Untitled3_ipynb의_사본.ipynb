```python
model = AutoModelForSeq2SeqLM.from_pretrained('/content/drive/MyDrive/fine_tuned_kobart')
tokenizer = AutoTokenizer.from_pretrained('/content/drive/MyDrive/fine_tuned_kobart_tokenizer')
```

```python
import re

# 영어 문장만 삭제하는 함수
def remove_english_sentences(text):
    # 영어만 포함된 문장 패턴 정의 (문장 끝에 . ! ? 등의 구두점이 오고, 영어 알파벳 및 숫자만 포함)
    result = re.sub(r'[A-Za-z0-9\s,;.!?\'"-]+(?:\s+|$)', '', text)
    return result

# 기사 내용 예시
article = wdb_df['text'][0]
# 영어 문장을 삭제
cleaned_article = remove_english_sentences(article)

# 길이를 1000자로 분할
article_parts = [cleaned_article[i:i+1000] for i in range(0, len(cleaned_article), 1000)]

# 각 부분에 대해 요약 생성
full_summary = ""  # 전체 요약을 담을 변수

for part in article_parts:
    # 입력 텍스트 토큰화
    inputs = tokenizer(part, return_tensors="pt", max_length=512, truncation=True, padding="max_length")

    # 모델을 통해 요약 생성
    summary_ids = model.generate(
        inputs["input_ids"],
        max_length=200,         # 각 요약의 최대 길이 설정
        min_length=50,          # 최소 길이 설정
        num_beams=6,            # 빔 서치 개수 조정 (다양한 후보를 고려)
        length_penalty=1.5,     # 길이를 더 선호하도록 설정
        early_stopping=True     # 빨리 멈추지 않도록 설정
    )

    # 요약을 디코딩하여 텍스트로 변환
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # 생성된 요약을 전체 요약에 추가
    full_summary += summary + " "  # 각 요약을 이어붙임

sum = []
sum.append(full_summary)

```
